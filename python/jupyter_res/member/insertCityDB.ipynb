```python
import csv
```

```python
f = open('../../mysql/cityDb.csv', 'r', encoding='UTF8')
csv_reader = csv.reader(f, delimiter=',')

print(csv_reader)
```

```python
cnt = 0

for row in csv_reader:
    if cnt == 0:
        cnt += 1
        continue
        
    print(row)
    city_no = (row[0])
    city = (row[1])
    district = (row[2])
    villageName = (row[3])
    
    print(city_no, city, district, villageName)

    sql = """
        insert into city (city_no, city, district, village_name) values ( %s, %s, %s, %s)
    """
    
    cursor.execute(sql, (city_no, city, district, villageName))
    
db.commit()

f.close()
db.close()
```
